```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1, num=200)
plt.plot(x,x)
plt.show()
```

```python
def membership(value, bottom_left, top_left, top_right, bottom_right):
    if value < bottom_left:
        return 0.0
    elif value < top_left:
        return (value-bottom_left) / (top_left-bottom_left)
    elif value < top_right:
        return 1.0
    elif value < bottom_right:
        return 1.0 - ((value-top_right) / (bottom_right-top_right))
    else:
        return 0.0

water_func     = np.vectorize(lambda x: membership(x, -1,0,0,0.1))
sand_func      = np.vectorize(lambda x: membership(x, 0.1, 0.2, 0.3, 0.4))
grass_func     = np.vectorize(lambda x: membership(x, 0.3, 0.4, 0.6, 0.7))
rockgrass_func = np.vectorize(lambda x: membership(x, 0.6, 0.7, 0.8, 0.9))
rock_func     = np.vectorize(lambda x: membership(x, 0.8, 0.9, 1.0, 2.0))
```

```python
plt.plot(x,water_func(x))
plt.plot(x,sand_func(x))
plt.plot(x,grass_func(x))
plt.plot(x,rockgrass_func(x))
plt.plot(x,rock_func(x))
plt.plot([0,1], [-0.1,1.1], c="white")
plt.show()
```
